**Task-3: **Prediction using Decision Tree Algorithm :
Create the Decision Tree classifier and visualize it graphically. The purpose is if we feed any new data to this classifier, it would be able to predict the right class accordingly.



```python
#Importing the required libraries
import pandas as pd
import seaborn as sns
```

```python
#Reading the dataset
names=['sepal-length','sepal-width','petal-length','petal-width','Class']
data=pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",names=names)

data
```

```python
#Dividing the dataset into dependent and independent variables
x = data.iloc[:,:4].values
y = data.iloc[:,4:5]
```

```python
#Spliting the data into train and test dataset
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.20, random_state=0)
```

```python

from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
# Accuracy of the model
print("Accuracy: ", metrics.accuracy_score(Y_test,y_pred))
```

```python
#predicting class for new data
new_data = [[4.6, 2.0, 4.2, 1.1]]
y_pred = model.predict(new_data)
print(y_pred)
```

```python
#vizualise decision tree
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize = (20,10))
tree.plot_tree(model,filled=True,rounded=True)
plt.show()
```
